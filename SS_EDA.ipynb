```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('C:/Users/Taj Hasan Mansuri/Downloads/Data analysis assignment.xlsx',sheet_name='Primary Sale Data')
```

```python
data1=data.dropna(axis=1)
```

```python
data1.info()
```

```python
data1.describe()
```

```python
data1.isna().sum().sum()
```

```python
# Histogram of sales quantity
plt.hist(data1['QUANTITY'], bins=10)
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.title('Distribution of Sales Quantity')
plt.show()
```

```python
# Bar chart of sales by month
sales_by_month = data1.groupby('Month')['QUANTITY'].sum()
sales_by_month.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Total Sales Quantity')
plt.title('Total Sales Quantity by Month')
plt.show()
```

```python
# Box plot of taxable amount by material description
plt.figure(figsize=(10, 6))
data1.boxplot(column='TAXABLE AMOUNT', by='MATERIAL DESC')
plt.xticks(rotation=90)
plt.xlabel('Material Description')
plt.ylabel('Taxable Amount')
plt.title('Distribution of Taxable Amount by Material Description')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = data1.corr()
plt.figure(figsize=(12, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(data.columns)), data.columns, rotation=90)
plt.yticks(range(len(data.columns)), data.columns)
plt.title('Correlation Matrix')
plt.show()
```

# Sales Distribution by Material Description:

```python
plt.figure(figsize=(10, 6))
data1['MATERIAL DESC'].value_counts().plot(kind='bar')
plt.xlabel('Material Description')
plt.ylabel('Frequency')
plt.title('Distribution of Sales by Material Description')
plt.xticks(rotation=90)
plt.show()
```

# Top Selling Materials:


```python
top_selling_materials = data1['MATERIAL DESC'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
top_selling_materials.plot(kind='bar')
plt.xlabel('Material Description')
plt.ylabel('Quantity Sold')
plt.title('Top 10 Selling Materials')
plt.xticks(rotation=90)
plt.show()
```

# Sales by city:

```python
plt.figure(figsize=(10, 6))
data1['SHIP TO PARTY CITY'].value_counts().plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.title('Sales by city')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
data1.groupby('Invoice Date')['QUANTITY'].sum().plot()
plt.xlabel('Invoice Date')
plt.ylabel('Total Sales Quantity')
plt.title('Sales Trend Over Time')
plt.show()
```

```python
# Average Sales Quantity by Month:
```

```python
monthly_sales = data1.groupby('Month')['QUANTITY'].mean()
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Average Sales Quantity')
plt.title('Average Sales Quantity by Month')
plt.xticks(rotation=90)
plt.show()
```

```python
# Sales by Transporter Name:
```

```python
plt.figure(figsize=(10, 6))
data1['TRANSPORTER NAME'].value_counts().plot(kind='bar')
plt.xlabel('Transporter Name')
plt.ylabel('Frequency')
plt.title('Sales by Transporter Name')
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select relevant columns for segmentation
customer_data = data1[['SOLD-TO-PARTY CODE', 'QUANTITY', 'TAXABLE AMOUNT']]

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data)

# Apply K-means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(scaled_data)

# Analyze customer segments
cluster_counts = customer_data['Cluster'].value_counts()
cluster_counts.plot(kind='bar')
plt.xlabel('Cluster')
plt.ylabel('Number of Customers')
plt.title('Customer Segmentation')
plt.show()
```
